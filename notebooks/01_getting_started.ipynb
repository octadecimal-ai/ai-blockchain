## 1. Pobieranie Danych z Binance


```python
# Inicjalizacja kolektora
collector = BinanceCollector(sandbox=False)

# Pobierz dane BTC/USDC (zmienione z BTC/USDT)
df = collector.fetch_ohlcv(
    symbol="BTC/USDC",
    timeframe="1h",
    limit=500
)

print(f"Pobrano {len(df)} świec")
df.tail()
```

## 2. Analiza Techniczna


```python
# Dodaj wskaźniki techniczne
analyzer = TechnicalAnalyzer(df)
analyzer.add_all_indicators()

# Pobierz DataFrame ze wskaźnikami
df_ta = analyzer.get_dataframe()

# Wyświetl sygnały
signals = analyzer.get_signals()
print("📊 SYGNAŁY TECHNICZNE:")
for indicator, signal in signals.items():
    print(f"  • {indicator.upper()}: {signal}")

df_ta.tail()
```
